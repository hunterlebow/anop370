```python
## Decision variables --> Companys
## Objective function --> 8.56*ACME + 9.5*DynaStar etc...
## Constraints --> Companyi <= 25%
    ## ACME + Sabre Systems + DynaStar + MM >= 50%
    ## DynaStar + Eagle Vision + OptiPro <= 35%
```

Summation over $x$ variables 1 to 6:
$\sum_{i=1}^6 x_i $

```python
import gurobipy as gb
model = gb.Model()

total_amount = 750000
longTermCompanies = [0, 1, 3, 5]
highRiskCompanies = [1, 2, 4]
returns = [0.0865, 0.095, 0.10, 0.0875, 0.0925, 0.09]

company = model.addVars(6, ub=0.25*total_amount)
model.addConstr(gb.quicksum(company[i] for i in range(7)) == total_amount)
model.addConstr(
    gb.quicksum(longTermCompanies[i] for i in range(7)) >= 0.5*total_amount
    )
    
model.addConstr(
    gb.quicksum(highRiskCompanies[i] for i in range(7)) >= 0.35*total_amount
    )


model.setObjective((returns[i]*company[i] for i in range(7)), gb.GRB.MAXIMIZE )

model.optimize()


```
