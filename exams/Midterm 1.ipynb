### Exercise 1 [50%]

Formulate Pisghetti's optimization problem using the same notation from class, write down and solve the model using gurobipy, and then clearly print the number of servings that would be made of each dish as well as the revenue.

# Write your formulation here!
<h3><b>Decision Variables</b></h3>
<ul>

- $Dish_0$ = Number of servings of Fifone to make</br>
- $Dish_1$ = Number of servings of Mangione to make</br>
- $Dish_2$ = Number of servings of Paffuto to make </br>
- $Dish_3$ = Number of servings of Ridarella to make</br>
</ul>
<h3><b>Objective Funciton</b></h3>
<ul>
Each serving of Fifone yields a $12 profit, each serving of mangione yields a $10 profit, each serving of paffuto yields a $18 profit, each serving of ridarella yields an $8 profit. The objective is to maximize profit.
</br>

- $\max(12Dish_0 + 20Dish_1 + 18Dish_2 + 8Dish_3)$</br></ul>

<h3><b>Constraints</b></h3>
<ul>

$Dish_0$ requires 0.7 pounds of pasta, $Dish_1$ requires 1.5 pounds of pasta, $Dish_2$ requires 1.2 pounds of pasta, $Dish_3$ requires 0.8 pounds of pasta. The maximum amount of pasta per day is 1800 pounds.
- $0.7Dish_0 + 1.5Dish_1 + 1.2Dish_2 + 0.8Dish_3 \leq 1800$</br>

$Dish_0$ requires 0.1 pounds of cheese, $Dish_1$ requires 0.4 pounds of cheese, $Dish_2$ requires 0.3 pounds of cheese, $Dish_3$ requires 0.1 pounds of cheese.  The maximum amount of cheese per day is 150 pounds.</br>
- $0.1Dish_0 + 0.4Dish_1 + 0.3Dish_2 + 0.1Dish_3 \leq 150$</br>

$Dish_0$ requires 0.2 pounds of sauce, $Dish_1$ requires 0.5 pounds of sauce, $Dish_2$ requires 0.7 pounds of sauce. The maximum amount of tomato sauce per day is 1100 pounds.</br>
- $0.2Dish_0 + 0.5Dish_1 + 0.7Dish_2  \leq 1100$</br>

$Dish_0$ requires 1 minute of labor, $Dish_1$ requires 3 minutes of labor, $Dish_2$ requires 3 minutes of labor, $Dish_3$ requires 2 minutes of labor.  With each chef having 8 hour shifts and there are 3 chefs, the total working hours are 24.  This equates to 1440 working minutes.</br>
- $Dish_0 + 3Dish_1 + 3Dish_2 + 2Dish_3 \leq 1440$</br>

Upper bounds given the demand of 1000 fifones, 200 mangiones, 300 paffutos, and 500 ridarellas
- $Dish_0 \leq 1000$</br>
- $Dish_1 \leq 200$</br>
- $Dish_2 \leq 300$</br>
- $Dish_3 \leq 500$</br>

Lower bound (nonnegative)
- $Dish_i \geq 0$</br>

```python
import gurobipy as gb

model1 = gb.Model()


TOTAL_DISHES = 4
DEMAND = [1000, 200, 300, 500]
dish = model1.addVars(TOTAL_DISHES, ub=DEMAND)

"""PASTA"""
MAX_PASTA = 1800
PASTA = [0.7, 1.5, 1.2, 0.8]
pasta_constr = model1.addConstr(gb.quicksum(dish[i]*PASTA[i] for i in range(TOTAL_DISHES)) <= MAX_PASTA)

"""CHEESE"""
MAX_CHEESE = 150
CHEESE = [0.1, 0.4, 0.3, 0.1]
cheese_contsr = model1.addConstr(gb.quicksum(dish[i]*CHEESE[i] for i in range(TOTAL_DISHES)) <= MAX_CHEESE)

"""SAUCE"""
MAX_SAUCE = 1100
SAUCE = [0.2, 0.5, 0.7, 0]
sauce_constr = model1.addConstr(gb.quicksum(dish[i]*SAUCE[i] for i in range(TOTAL_DISHES)) <= MAX_SAUCE)

"""LABOR"""
HOURS_PER_CHEF = 8
CHEFS = 3
MAX_MINUTES = (HOURS_PER_CHEF*CHEFS)*60
LABOR  = [1, 3, 3, 2]
labor_constr = model1.addConstr(gb.quicksum(dish[i]*LABOR[i] for i in range(TOTAL_DISHES)) <= MAX_MINUTES)

"""OBJECTIVE FUNCTION - MAXIMIZE PROFIT"""
PROFIT = [12, 20, 18, 8]
model1.setObjective(gb.quicksum(dish[i]*PROFIT[i] for i in range(TOTAL_DISHES)), gb.GRB.MAXIMIZE)

model1.optimize()
```

```python
DISH_NAMES = ["Fifone", "Mangione", "Paffuto", "Ridarella"]

if model1.Status == gb.GRB.Status.OPTIMAL:
    for v in model1.getVars():
        print(f"{DISH_NAMES[v.index]}: {v.X} servings")
    print(f"\nModel 1 Total Profit: ${model1.ObjVal}")
```

```python
# Your Python code goes here!

model2 = gb.Model()

TOTAL_DISHES = 5
DEMAND = [1000, 200, 300, 500, 1800]
dish = model2.addVars(TOTAL_DISHES, ub=DEMAND)

"""PASTA"""
MAX_PASTA = 1800
PASTA = [0.7, 1.5, 1.2, 0.8, 1]
pasta_constr = model2.addConstr(gb.quicksum(dish[i]*PASTA[i] for i in range(TOTAL_DISHES)) <= MAX_PASTA)

"""CHEESE"""
MAX_CHEESE = 150
CHEESE = [0.1, 0.4, 0.3, 0.1, 0]
cheese_contsr = model2.addConstr(gb.quicksum(dish[i]*CHEESE[i] for i in range(TOTAL_DISHES)) <= MAX_CHEESE)

"""SAUCE"""
MAX_SAUCE = 1100
SAUCE = [0.2, 0.5, 0.7, 0, 0]
sauce_constr = model2.addConstr(gb.quicksum(dish[i]*SAUCE[i] for i in range(TOTAL_DISHES)) <= MAX_SAUCE)

"""LABOR"""
HOURS_PER_CHEF = 8
CHEFS = 3
MAX_MINUTES = (HOURS_PER_CHEF*CHEFS)*60
LABOR  = [1, 3, 3, 2, 0]
labor_constr = model2.addConstr(gb.quicksum(dish[i]*LABOR[i] for i in range(TOTAL_DISHES)) <= MAX_MINUTES)

"""OBJECTIVE FUNCTION - MAXIMIZE PROFIT"""
PROFIT = [12, 20, 18, 8, 2]
model2.setObjective(gb.quicksum(dish[i]*PROFIT[i] for i in range(TOTAL_DISHES)), gb.GRB.MAXIMIZE)

model2.optimize()
```

```python
DISH_NAMES = ["Fifone", "Mangione", "Paffuto", "Ridarella", "Pasta Dough Not Prepared"]
PASTA_DOUGH = 4


if model2.Status == gb.GRB.Status.OPTIMAL:
    for v in model2.getVars():
        if v.index < PASTA_DOUGH:
            unit = "servings"
        else:
            unit = "pounds"
        print(f"{DISH_NAMES[v.index]}: {v.X} {unit}")
    print(f"\nMoney saved from producing {v.X} lbs less of pasta: ${v.X * PROFIT[PASTA_DOUGH]}")
    print(f"\nModel 2 Total Profit: ${model2.ObjVal}")
```
